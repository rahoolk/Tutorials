```python
from urllib.request import urlretrieve
urlretrieve(URL,'Fremont.csv')
```

```python
import pandas as pd
data= pd.read_csv('Fremont.csv', index_col='Date', parse_dates=True)
data.head()
```

```python
%matplotlib inline
data.plot()
```

```python
#total number of bikes
%matplotlib inline
data.resample('W').sum().plot();
```

```python
import matplotlib.pyplot as plt
plt.style.use('seaborn')
data.columns= ['West', 'East']
data.resample('W').sum().plot();
```

```python
#Resample daily
data['Total']=data['West']+data['East']
ax= data.resample('D').sum().rolling(365).sum().plot();
ax.set_ylim(0,None); #Better idea of the change, not being so dramatic
```

```python
#Take a look at the trend within individual days
#groupby the time of day and take the mean and plot it
#We see that the eastbound sidewalk peaks in the afternoon, and westbound peaks in the morning

data.groupby(data.index.time).mean().plot();
```

```python
#Make a pivot table
pivoted= data.pivot_table('Total', index=data.index.time, columns=data.index.date)
pivoted.iloc[:5,:5] #look at first 5x5 set of data in the full set
```

```python
#Have a line for each day of the four years. 
pivoted.plot(legend=False);
```

```python
#Transparent lines- get an idea of how trend in crossing during the day changes over four year period
#the commute days are weekdays, but broad-usage days are weekends
pivoted.plot(legend=False, alpha=0.01);
```
