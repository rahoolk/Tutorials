# Mixed Power Sources for Electricity Generation

Following Ben Walsh's tutorial on how to generate stacked area charts using Python's Altair, and how to compare  net consumption (TBtu) from different fuel sources over the past 50 years. This data comes from the Open EI project, originally from the following: https://openei.org/datasets/dataset/energy-intensity-indicators-in-the-u-s-electricity-sector-1949-2004

```python
import altair as alt
alt.renderers.enable('notebook')
```

```python
import pandas as pd
data= pd.read_csv('net_consumption.csv')
```

```python
data.head()
```

```python
alt.Chart(data).mark_area().encode(
    x="year",
    y="net_consumption_Tbtu:Q",
    color="source:N"
)
```

```python
alt.Chart(data).mark_area(opacity=0.3).encode(
    x="year:T",
    y=alt.Y("net_consumption_Tbtu:Q", stack=None),
    color="source:N"
)
```

```python
alt.Chart(data).mark_area().encode(
    x="year:T",
    y="net_consumption_Tbtu:Q",
    color="source:N",
    row="source:N"
)
```

```python
import pandas as pd
data.year = pd.to_datetime(data.year, format="%Y")
alt.Chart(data).mark_area().encode(
    x="year:T",
    y="net_consumption_Tbtu:Q",
    color="source:N"
)
```
