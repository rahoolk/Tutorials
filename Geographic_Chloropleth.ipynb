```python
#IOU Electricity Rates by Zipcode
#https://openei.org/doe-opendata/dataset/u-s-electric-utility-companies-and-rates-look-up-by-zipcode-2014/resource/c3acb9a5-bb18-462c-ad2b-6fd6136d6f98
import numpy as np
import pandas as pd
import plotly.plotly as py
import plotly.tools as tls
```

```python
#create an account on plot.ly and they will generate an api key for you.
tls.set_credentials_file(username='',api_key='')
```

```python
import urllib  # the module for reading a url
import csv  # the module for opening a .csv format file



# read a .xlsx format file with a given URL

from urllib.request import urlopen
import matplotlib.pyplot as plt

 
url = 'http://en.openei.org/doe-opendata/dataset/1f0a98f8-b221-40cb-b6d6-f1a6df12ee0a/resource/c3acb9a5-bb18-462c-ad2b-6fd6136d6f98/download/iouzipcodes2014.csv'
rates_csv = urlopen(url)
 
df = pd.read_csv(rates_csv, index_col=0, parse_dates=True)
```

```python
df.head()
```

```python
states= df.groupby('state').mean().reset_index()
```

```python
states.head()
```

```python
states.columns=['state','not_applicable','community','industry','residential']
states=states.round(2)
```

```python
states.head()
```

```python
states['text']= 'State: '+ states['state']+ '<br>'+'Residential_Rate: '+states['residential'].astype(str) +  '¢/kWh'+'<br>'+ 'Industry_Rate: ' + states['industry'].astype(str) +  '¢/kWh'+'<br>'+ 'Community_Rate : ' + states['community'].astype(str)+ '¢/kWh' 

data = [ dict(
        type='choropleth',
        autocolorscale = False,
        locations = states['state'].astype(str),
        z = states['residential'].astype(float),
        locationmode = 'USA-states',
        text = states['text'],
        marker = dict(
            line = dict (
                color = 'rgb(255,255,255)',
                width = 2
            ) ),
        colorbar = dict(
            title = "Cost ¢/kWh")
        ) ]


data
```

```python
layout= dict(title='Investor Owned Utilities Residential Electricity Rates',
            geo=dict(scope='usa', projection=dict(type='albers usa'), showlakes= True, lakecolor= 'rgb(66,165,245)',),)
layout
```

```python
fig= dict(data=data, layout=layout)
py.iplot(fig, filename='electricity_rate')
```
